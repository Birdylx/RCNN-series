## RoI Pooling 

```python
import torchvision
import torch
import torchvision.transforms as T
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
# read image 
img = plt.imread("human.png")
plt.imshow(img)
plt.show()
img.shape
```

```python
# assume dowansample stride is 16
feature_map = torch.randn((1,256,int(1027/16),int(1288/16)))
feature_map.shape
```

```python
# assume we have a proposal, this also the rpn output
proposal = torch.tensor([717.0974, 141.1179, 869.0415, 288.6204])
proposal

# plot proposal on image
box = proposal.numpy()
img = cv2.rectangle(img,(box[0],box[1]),(box[2],box[3]),color=(255,255,0),thickness=3)
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img)
plt.show()
```

```python
# get roi coordinates on feature map
roi_on_feature = proposal/16.
roi_on_feature
```

```python
# roi pool on the feature map to fixed size=7
# we need to cut the roi area on feature map into 7 bins
# first, ceil the roi coordinates, this introduces the cutoff error
roi_on_feature = torch.ceil(roi_on_feature).numpy().astype(np.int)
roi_on_feature
```

```python
import math
start_x = roi_on_feature[0]
start_y = roi_on_feature[1]

roi_w = roi_on_feature[2]-roi_on_feature[0]
roi_h = roi_on_feature[3]-roi_on_feature[1]
bin_size_w = roi_w/7
bin_size_h = roi_h/7

result = torch.zeros((7,7))

for i in range(7):
    for j in range(7):
        bin = feature_map[0,0,start_y+math.floor(i*bin_size_h):start_y+math.ceil((i+1)*bin_size_h),start_x+math.floor(j*bin_size_w):start_x+math.ceil((j+1)*bin_size_w)]
        max_value = bin.max()
        result[i,j] = max_value
result
```

```python
# check out the pytorch implementation result, totaly the same
from torchvision.ops import roi_pool
rois_in_feature = torch.tensor([0,44.8186,8.8199, 54.3151, 18.0388])
result = roi_pool(feature_map,rois_in_feature,output_size=(7,7))
result[0][0]
```
